**Import pakietów:**

````julia
using Polynomials
using Plots
using QuadGK
using Printf
````

**Implementacja wielomianów Legendre'a zgodnie ze wzorem:**

````julia
function LegendrePolynomials(x)
    if x == 0
        return Polynomial([1])
    end
    
    if x == 1
        return Polynomial([0, 1])
    end
    
    return (2 * x - 1) / (x) * LegendrePolynomials(1) * LegendrePolynomials(x-1) - (x - 1) / (x) * LegendrePolynomials(x - 2)
end
````

**Wykres:**

````julia
xs = -1:0.001:1
ys = [0(x) for x in xs]

plot_1 = scatter(0, 0)

for i in 0:4
    ys = [LegendrePolynomials(i)(x) for x in xs]
    plot_1 = plot!(xs, ys, label="$(i)")
end

plot_1
````

````julia
for i in 2:4
    @printf("\nRoots:\t%s\n", roots(LegendrePolynomials(i)))
    @printf("Gauss:\t%s\n", gauss(i)[1])
end
````

**Implementacja funkcji:**

````julia
function integralG(k, f)
    (xp, a) = gauss(Float64, k)
    sum(a .* f.(xp))
end
````

### Zadanie 3 (1p)
  
Skorzystaj z rozwiązania zadania 2 do napisania funkcji 
liczącej  całki w dowolnym przedziale $\int_{a}^{b} f(x) dx$ 

dokonując normalizacji  do $\int_{-1}^{1} F(z) dz$ 

podstawiając:

$x=\frac{b+a}{2}+ \frac{b-a}{2} z $ oraz 

$dx =\frac{b-a}{2} dz $

Przetestuj działanie na kilku przykładach i sprawdź z wynikami otrzymanymi analitycznie.


**Implementacja:**

````julia
function integralG_ab(k, f, a, b)
    (xp, w) = gauss(Float64, k)
    sum(w .* f.((((b + a) / 2).+ ((b - a) / 2).* xp)).* ((b - a) / 2)) 
end
````

````julia
for i in 1:4
    f_(x) = x^i
    F_(x) = 1/(i + 1) * x^(i + 1)
    
    println("\nFunction no $i:\n", integralG_ab(i, f_, 0, 10))
    println("Compare to:\n", F_(10) - F_(0))

end
````

### Zadanie 5 (3p)
Napisz własne funkcje całkujące -  w wybranym przedziale (a,b) - złożonymi metodami prostokątów, trapezów oraz Simpsona. Dla stałego przedziału zwiększaj liczbę użytych punktów (n). Narysuj wykresy funkcji błędu  w stosunku do wyniku otrzymanego analitycznie, 
w zależnosci od $h=\frac{b-a}{n}$:
1. wykładniczej
2. trygonometrycznej (np. sin (x) w przedziale $[0,2\pi]$)


**Metoda prostokątów:**

````julia
function integral_rectangle(k, f, a, b)
    h = (b - a) / k
    result = 0
    for i in 1:k-1
        result += f(a + i * h) * h
    end
    return result
end
````

**Metoda trapezów:**

````julia
function integral_trapezoidal(k, f, a, b)
    # Calculate step size.
    h = (b - a) / k
    # Find sum.
    result = f(a) + f(b)
    
    for i in 1:k
        m = a + i * h
        result += 2 * f(m)
    end
    result = result * (h / 2)
    return result
end
````

**Metoda Simpsona:**

````julia
function integral_simpson(k, f, a, b)
    # Calculate step size.
    h = (b - a) / k
    # Find sum.
    result = f(a) + f(b)
    
    for i in 1:k
        m = a + i * h
        if i%2 == 0
            result += 2 * f(m)
        else
            result += 4 * f(m)
        end
    end
    result = result * (h / 3)
    return result
end
````

````julia
f_(x) = x^4
````

````julia
integral_rectangle(1000, f_, 0, 10)
````

````julia
integral_trapezoidal(1000, f_, 0, 10)
````

````julia
integral_simpson(1000, f_, 0, 10)
````
